```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import *
from nltk.stem.porter import *
nltk.download('stopwords')
nltk.download('punkt')
```

# **Create a Data Frame for the dataset**

```python
encoding= 'latin-1'
df = pd.read_csv("/content/Songs-Search-engine/allsongs_data.csv", usecols=["songID","Album", "Lyrics", "Artist", "Song Title", "Year"], nrows=6050,encoding= encoding)
```

```python
lyrics = df['Lyrics']

song_titles = df['Song Title']
from nltk.tokenize import sent_tokenize

for lyric, title in zip(lyrics, song_titles):
  sentences = sent_tokenize(lyric)
  all_tokens = []
  for sentence in sentences:
    tokens = word_tokenize(sentence)
    all_tokens.extend(tokens)
```

# **Remove the sopwords from the Lyrics column and make all of the text in lower case**

```python
music_stopwords =  ["oh", "yeah", "mm", "la"]
section_stopwords =  ["intro", "verse", "chorus", "bridge", "outro" , "shhhhhhhhh", "luke"]

stop_words = stopwords.words('english') + music_stopwords + section_stopwords

def remove_stopwords(text):
    if isinstance(text, str):
        tokens = word_tokenize(text)
        filtered_tokens = [word.lower() for word in tokens if word.lower() not in stop_words]
        return ' '.join(filtered_tokens)
    else:
        return ''

df['processed_Lyrics']=df['Lyrics'].str.lower()
df['processed_Lyrics'] = df['processed_Lyrics'].apply(remove_stopwords)
```

```python
def remove_punctuation(text):
    return re.sub(r'[^\w\s]', '' , text)
df['processed_Lyrics'] = df['processed_Lyrics'].apply(remove_punctuation)
```

# **Apply Stemming on the Lyrics column**

```python
stemmer = PorterStemmer()

def Steem_text(text):

    tokens = word_tokenize(text)
    stemmed_tokens = [stemmer.stem(word) for word in tokens]
    return ' '.join(stemmed_tokens)

df['processed_Lyrics']=df['processed_Lyrics'].apply(Steem_text)
```

```python
inverted_index = {}

for song_id, processed_lyrics in zip(df['songID'], df['processed_Lyrics']):
    tokens = processed_lyrics.split()
    unique_terms = set(tokens)
    term_freq = {}
    for term in tokens:
        term_freq[term] = term_freq.get(term, 0) + 1

    for term in unique_terms:
        if term not in inverted_index:
            inverted_index[term] = {}
        inverted_index[term][song_id] = term_freq[term]


for term, postings in inverted_index.items():
    print(f"{term} -> {postings}")
```

```python
def preprocess_query(query):
    tokens = query.lower().split()
    tokens = [remove_punctuation(token) for token in tokens if token not in stop_words]
    return tokens


def search_with_batch_retrieve(user_query, index_path):
    retr = pt.BatchRetrieve(index_path, controls={"wmodel": "TF_IDF"}, num_results=5)
    res = retr.search(user_query)
    return res

def rank_documents_with_batch_retrieve(user_query, relevant_docs, df_reset):
    ranked_results = []
    for i, doc in enumerate(relevant_docs.iterrows()):
        doc_id = doc[1]['docno']
        score = doc[1]['score']
        matching_rows = df_reset[df_reset['docno'] == doc_id]
        if not matching_rows.empty:
            title = matching_rows['Song Title'].iloc[0]
            year = matching_rows['Year'].iloc[0]
            artist = matching_rows['Artist'].iloc[0]
            album = matching_rows['Album'].iloc[0]

            ranked_results.append((i + 1, doc_id, score, title, year, album, artist))
        else:
            print(f"No matching rows found for doc_id: {doc_id}. Skipping.")
    return ranked_results


df_reset = df.reset_index()
user_query = "halo girlfriend"
index_path = '/content/myFirstIndex/data.properties'
relevant_docs = search_with_batch_retrieve(user_query, index_path)

ranked_results = rank_documents_with_batch_retrieve(user_query, relevant_docs, df_reset)

for rank, doc_id, score, title, year, album, artist in ranked_results:
    print(f"Rank: {rank}, Document ID: {doc_id}, Score: {score}")
    print(f"Title: {title}, Year: {year}, Album: {album}, Artist: {artist}")
    print()
```

```python
import torch
from sklearn.metrics.pairwise import cosine_similarity
from transformers import BertModel, BertTokenizer
```

```python
import os


bert_model = BertModel.from_pretrained('bert-base-uncased')
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
def perform_query_expansion(user_query, index_path):
    bert_model = BertModel.from_pretrained('bert-base-uncased')
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

    relevant_docs = search_with_batch_retrieve(user_query, index_path)

    bm25 = pt.BatchRetrieve(index_path, wmodel="BM25", num_results=5)
    rm3_expander = pt.rewrite.RM3(index_path, fb_terms=10, fb_docs=100)

    rm3_qe = bm25 >> rm3_expander

    search_result = rm3_qe.search(user_query)
    if not search_result.empty:
        expanded_query = search_result.iloc[0]["query"]
    else:
        expanded_query = ""

    query_embedding = get_bert_embedding(user_query, bert_model)
    top_doc_embeddings = [get_bert_embedding(doc, bert_model) for doc in relevant_docs[:5]]

    related_terms = find_similar_terms(query_embedding, top_doc_embeddings)

    print("Expanded Terms:")
    for term in expanded_query.split()[1:]:
        print(term)

    expanded_query_formatted = ' '.join(expanded_query.split()[1:])

    results_wqe = bm25.search(expanded_query_formatted)
    docids2 = results_wqe['docid'].values.tolist()
    scores2 = results_wqe['score'].values.tolist()
    docids1 = relevant_docs['docid'].values.tolist()
    scores1 = relevant_docs['score'].values.tolist()

    print("\n Before Expansion \t After Expansion")
    print(pd.concat([relevant_docs[['docid', 'score']][0:5].add_suffix('_1'),
                     results_wqe[['docid', 'score']][0:5].add_suffix('_2')],
                    axis=1).fillna(''))

    if os.path.exists('expanded_query.txt'):
        os.remove('expanded_query.txt')

    with open('expanded_query.txt', 'w') as f:
        f.write("Expanded Terms:\n")
        for term in expanded_query.split()[1:]:
            f.write(term + "\n")

        f.write("\nBefore Expansion:\n")
        for docid, score in zip(relevant_docs['docid'], relevant_docs['score']):
            f.write(f"DocID: {docid}, Score: {score}\n")

        f.write("\nAfter Expansion:\n")
        for docid, score in zip(results_wqe['docid'], results_wqe['score']):
            f.write(f"DocID: {docid}, Score: {score}\n")

    return expanded_query_formatted, related_terms, expanded_query

def get_bert_embedding(text, bert_model):
    encoded_input = tokenizer(text, return_tensors='pt')
    with torch.no_grad():
        output = bert_model(**encoded_input)
    return output.pooler_output.squeeze(0)

def find_similar_terms(query_embedding, doc_embeddings, threshold=0.8, top_k=5):
    similarities = cosine_similarity(query_embedding.reshape(1, -1), doc_embeddings)[0]
    sorted_indices = np.argsort(similarities)[::-1]
    similar_terms = []
    for idx in sorted_indices[:top_k]:
        if similarities[idx] >= threshold:
            similar_terms.append(idx)
    return similar_terms

user_query = "someday, when my life has passed"
index_path = "/content/myFirstIndex/data.properties"
expanded_query = perform_query_expansion(user_query, index_path)
```

```python
import pandas as pd

def read_expanded_query(file_path):
    expanded_terms = []
    before_expansion = []
    after_expansion = []

    with open(file_path, 'r') as f:
        lines = f.readlines()
        section = None
        for line in lines:
            line = line.strip()
            if line.startswith("Expanded Terms:"):
                section = "expanded_terms"
            elif line.startswith("Before Expansion:"):
                section = "before_expansion"
            elif line.startswith("After Expansion:"):
                section = "after_expansion"
            elif line:
                if section == "expanded_terms":
                    expanded_terms.append(line)
                elif section == "before_expansion" or section == "after_expansion":
                    parts = line.split(',')
                    docid = parts[0].split(':')[1].strip()
                    score = parts[1].split(':')[1].strip()
                    if section == "before_expansion":
                        before_expansion.append((docid, score))
                    else:
                        after_expansion.append((docid, score))

    before_df = pd.DataFrame(before_expansion, columns=['DocID', 'Score'])
    after_df = pd.DataFrame(after_expansion, columns=['DocID', 'Score'])

    return expanded_terms, before_df, after_df

# Usage example:
file_path = 'expanded_query.txt'
expanded_terms, before_expansion, after_expansion = read_expanded_query(file_path)

print("Expanded Terms:")
for term in expanded_terms:
    print(term)

print("\nBefore Expansion:")
print(before_expansion)

print("\nAfter Expansion:")
print(after_expansion)
```

# **Evaluation**

```python
import pandas as pd
import time

def evaluate_queries_with_time_and_precision(queries, df_expected_results, index_path):
    evaluation_results = []

    for query in queries:
        result = {}
        result['query'] = query

        start_time = time.time()

        relevant_docs = search_with_batch_retrieve(query, index_path)
        ranked_results = rank_documents_with_batch_retrieve(query, relevant_docs, df_reset)

        end_time = time.time()
        elapsed_time = end_time - start_time

        actual_docs = [doc_id for _, doc_id, _, _, _, _, _ in ranked_results]
        expected_docs = df_expected_results[df_expected_results["Query"] == query]["Expected_Relevant_Docs"].iloc[0]
        relevant_intersection = len(set(actual_docs) & set(expected_docs))
        precision = relevant_intersection / len(actual_docs) if len(actual_docs) > 0 else 0

        result['actual_docs'] = actual_docs
        result['precision'] = precision
        result['elapsed_time'] = elapsed_time

        evaluation_results.append(result)

    return evaluation_results

query_expected_results = [
    ("halo girlfriend", ["1"]),
    ("lay around wonder why always there", ["2"]),
    ("fly fly away", ["6"]),
    ("dream come true found you", ["7"]),
    ("panini meanie", ["8", "9"]),
    ("got what I got miss what I had", ["10"]),
    ("hello from the other side", ["20", "21"]),
    ("send my love to your new lover", ["22"]),
    ("when we were young", ["23"]),
    ("water under the bridge", ["24"]),
    ("only you", ["25", "26"]),
    ("intuition", ["27"]),
    ("where my girls at", ["28"]),
    ("bartender", ["29"]),
    ("cupid", ["26"])
]

df_queries_expected = pd.DataFrame(query_expected_results, columns=["Query", "Expected_Relevant_Docs"])

evaluate_queries_with_time_and_precision(df_queries_expected["Query"], df_queries_expected, index_path)
```

```python
from flask import Flask, render_template, request, redirect, url_for
from flask_ngrok import run_with_ngrok

app = Flask(__name__, template_folder='/content/Songs-Search-engine/GUI/templates', static_folder='/content/Songs-Search-engine/GUI/static')
run_with_ngrok(app)

@app.route("/")
def home():
    return render_template("main.html")

@app.route("/search", methods=['POST', 'GET'])
def search():
    user_query = request.form.get('query')
    if not user_query:
        return render_template("main.html", error="Please enter a query.")

    relevant_docs = search_with_batch_retrieve(user_query, index_path)
    ranked_results = rank_documents_with_batch_retrieve(user_query, relevant_docs, df_reset)

    expanded_query = perform_query_expansion(user_query, index_path)

    return render_template("main.html", query=user_query, results=ranked_results)

@app.route('/query_expansion', methods=['GET', 'POST'])
def query_expansion():
    file_path = 'expanded_query.txt'
    expanded_terms, before_expansion, after_expansion = read_expanded_query(file_path)
    before_expansion = before_expansion.to_dict('records')
    after_expansion = after_expansion.to_dict('records')
    return render_template('main.html', expanded_terms=expanded_terms, before_expansion=before_expansion, after_expansion=after_expansion)

@app.route('/evaluation', methods=['GET', 'POST'])
def evaluation():

    evaluation_results = evaluate_queries_with_time_and_precision(df_queries_expected["Query"], df_queries_expected, index_path)

    return render_template('evaluation.html', evaluation_results=evaluation_results)

if __name__ == "__main__":
    app.run()
```
